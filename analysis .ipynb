```python
import pandas as pd
df = pd.read_csv('sales_data.csv')
df
```

```python
df.head()
df.info()
```

```python
df.isnull().sum()
```

```python
total_revenue = df["Total_Sales"].sum()
sales_by_product = df.groupby("Product")["Total_Sales"].sum()
sales_by_region = df.groupby("Region")["Total_Sales"].sum()
best_product = sales_by_product.idxmax()
print("TOTAL REVENUE",total_revenue)
print("SALES BY PRODUCT",sales_by_product)
print("SALES BY REGION",sales_by_region)
print("BEST PRODUCT",best_product)
```

```python
import matplotlib.pyplot as plt
product_sales = df.groupby("Product")["Total_Sales"].sum()
plt.figure()
plt.bar(product_sales.index, product_sales.values)
plt.title("Total Sales by product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
region_sales = df.groupby("Region")["Total_Sales"].sum()
plt.figure()
plt.pie(region_sales.values, labels=region_sales.index)
plt.title("Sales Distribution by Region")
plt.show()
```
